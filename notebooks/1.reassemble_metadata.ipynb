# Reassemble metadata

This notebook rebuilds `metadata_selected.csv` from the resubmission source tables.

- Use the same `parnet_clean` kernel as the other revision notebooks.
- The notebook exposes the input/output paths up front, validates them, then runs the assembly through a helper module.


```python
from pathlib import Path
import importlib.util
import sys
import pandas as pd

REPO_ROOT = Path.cwd()
PIPELINE_PATH = REPO_ROOT / "resubmission" / "scripts" / "reassemble_metadata_pipeline.py"

spec = importlib.util.spec_from_file_location("reassemble_metadata_pipeline", PIPELINE_PATH)
meta_pipe = importlib.util.module_from_spec(spec)
assert spec.loader is not None
spec.loader.exec_module(meta_pipe)

print(f"Kernel python: {sys.executable}")
print(f"Pipeline module: {PIPELINE_PATH}")
```

## Controls

Override any input path here if you want to rebuild from a different table. In the common case you can leave everything as-is.


```python
INPUT_OVERRIDES = {
    "base_metadata": REPO_ROOT / "resubmission/data/combined_dataTPMno100.csv",
    "introns_source": REPO_ROOT / "resubmission/data/all_vastdb_introns_source.csv",
    "repeat_masker": REPO_ROOT / "resubmission/data/all_vastdb_introns_repeatmasker_scores.csv",
    "phylop": REPO_ROOT / "resubmission/data/all_vastdb_introns_phylop_scores.csv",
    "ssdrip_peaks": REPO_ROOT / "resubmission/data/all_vastdb_introns_ssDRIP_peaks.csv",
    "ssdrip_reads": REPO_ROOT / "resubmission/data/all_vastdb_introns_ssDRIP_reads.csv",
    "hl_gene": REPO_ROOT / "resubmission/data/hl/hl_genewise_filtered.csv",
    "hl_intron": REPO_ROOT / "resubmission/data/hl/hl_intronwise_filtered.csv",
    "speckles": REPO_ROOT / "resubmission/data/external/speckles_enrichment.csv",
    "cpg": REPO_ROOT / "resubmission/data/all_vastdb_introns_source_cpg_methylation.csv",
}

OUTPUT_OVERRIDES = {
    "metadata_selected": REPO_ROOT / "resubmission/data/metadata_selected.csv",
}

SAVE_OUTPUT = True
```

```python
CONFIG = meta_pipe.build_config(inputs=INPUT_OVERRIDES, outputs=OUTPUT_OVERRIDES)
pd.Series({key: str(value) for key, value in CONFIG.items()})
```

## Save

Only run this cell when you are happy with the assembled table.


```python
if SAVE_OUTPUT:
    meta_pipe.save_selected_metadata(tab_metadata_selected, CONFIG["metadata_selected"])
    print(f"Saved to {CONFIG['metadata_selected']}")
else:
    print("SAVE_OUTPUT is False; no file written.")
```

## Compare with existing file

This is useful when you want to confirm that a rerun reproduces the current `metadata_selected.csv`.


```python
existing_path = CONFIG["metadata_selected"]
if existing_path.exists():
    existing = pd.read_csv(existing_path, index_col=0)
    comparison = pd.Series({
        "same_shape": existing.shape == tab_metadata_selected.shape,
        "same_columns": list(existing.columns) == list(tab_metadata_selected.columns),
        "same_index": existing.index.equals(tab_metadata_selected.index),
    })
    print(comparison)
else:
    print(f"No existing file at {existing_path}")
```
